```python
from scipy import io
```

```python
allsplit_mat = io.loadmat("./docker_volume/sunrgbd/OFFICIAL_SUNRGBD/SUNRGBDtoolbox/traintestSUNRGBD/allsplit.mat")
```

```python
N_train = allsplit_mat["alltrain"].size
N_val = allsplit_mat["alltest"].size
```

```python
hash_train = {}
hash_val = {}
```

```python
for i in range(N_train):
    folder_path = allsplit_mat["alltrain"][0][i][0]
    folder_path = folder_path[16:]
    hash_train[folder_path] = 0
    
for i in range(N_val):
    folder_path = allsplit_mat["alltest"][0][i][0]
    folder_path = folder_path[16:]
    hash_val[folder_path] = 0
```

```python
hash_train
```

```python
meta3dbb = io.loadmat("./docker_volume/sunrgbd/OFFICIAL_SUNRGBD/SUNRGBDMeta3DBB_v2.mat")
```

```python
fid_train = open("./docker_volume/sunrgbd/sunrgbd_trainval/train_data_idx.txt", "w")
fid_val = open("./docker_volume/sunrgbd/sunrgbd_trainval/val_data_idx.txt", "w")
```

```python
SUNRGBDMeta = meta3dbb["SUNRGBDMeta"][0]
```

```python
meta3dbb["SUNRGBDMeta"]
```

```python
for imageId in range(10335):
    # print(imageId, end=" - ")
    data = SUNRGBDMeta[imageId]
    
    single_meta_dict = {
        "sequenceName": data["sequenceName"][0],
        "Rtilt": data["Rtilt"][0],
        "K": data["K"][0],
        "depthpath": data["depthpath"][0],
        "rgbpath": data["rgbpath"][0],
        "anno_extrinsics": data["anno_extrinsics"][0],
        "depthname": data["depthname"][0],
        "rgbname": data["rgbname"][0],
        "sensorType": data["sensorType"][0],
        "valid": data["valid"][0],
        "groundtruth3DBB": data["groundtruth3DBB"]
    }
    
    depth_path = single_meta_dict["depthpath"][16:]
    filepath = "/".join(depth_path.split("/")[:-2])
    # print(filepath, end=" - ")
    if filepath in hash_train:
        # print("train")
        fid_train.write(str(imageId) + "\n")
    elif filepath in hash_val:
        # print("val")
        fid_val.write(str(imageId) + "\n")
    else:
        print("Error")
```

```python
fid_train.close()
fid_val.close()
```
